```python
def colorRedBlue(answer):
    colors=[]
    for a in answer:
        if a==1:
            colors.append('red')
        else:
            colors.append('blue')
    return colors
```

```python
from matplotlib import pyplot as plt
x = df['Dias']
y = df['Visualizações']
cols=colorRedBlue(df['Pergunta'])
plt.scatter(x,y)
plt.title('Visualizações x Dias')
plt.xlabel('Dias')
#plt.gca().invert_xaxis()
plt.ylabel('Visualizações')
plt.scatter(x=x,y=y,c=cols)
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"g--")
plt.legend(('Views x Dias','Não pergunta','Pergunta'))
plt.show()
```

```python
def testMethod(X, method):
    lista_MSE = []
    lista_R2 = []
    t=time.perf_counter()

    for i in range (10):
        x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
        reg = method
        reg.fit(x_train, y_train)
        y_pred = reg.predict(x_test)
        lista_MSE.append(mean_squared_error(y_test, y_pred))
        lista_R2.append(r2_score(y_test, y_pred))
        
    pred = reg.predict(X)
    print("MSE médio: ", np.median(lista_MSE))
    print("R2 médio: ", np.median(lista_R2))
    print(round(time.perf_counter()-t,3), "segundos")
```

```python
dados = df
#dados = x
```

```python
parameters = {#'max_depth':[5, 10, 15, 20, 25, 30, 35], 
              'n_estimators':[10, 50, 100, 150, 200, 250, 300], 
              #'min_samples_split':[2, 5, 10, 20],
              'min_samples_leaf':[1, 5, 10, 15, 20]}

x_train, x_test, y_train, y_test = train_test_split(dados, y, test_size=0.25)
reg = RandomForestRegressor()
gs = GridSearchCV(reg, parameters)#, cv=10)
gs.fit(x_train, y_train)
y_pred = gs.predict(x_test)
print(gs.best_estimator_)


for i in range (2,len(df.columns)):
    print("\nPCA:",i)
    dadosPCA = PCA(n_components=i).fit_transform(dados)
    testMethod(dadosPCA,gs.best_estimator_)
    
print("\nNO PCA")
testMethod(dados,gs.best_estimator_)
```

```python
# kernel linear
parameters = {'epsilon':[0.05, 0.1, 0.15, 0.2, 0.25]}

x_train, x_test, y_train, y_test = train_test_split(dados, y, test_size=0.25)
reg = SVR(kernel='linear')
gs = GridSearchCV(reg, parameters)#, cv=10)
gs.fit(x_train, y_train)
y_pred = gs.predict(x_test)
print(gs.best_estimator_)

for i in range (2,len(df.columns)):
    print("\nPCA:",i)
    dadosPCA = PCA(n_components=i).fit_transform(dados)
    testMethod(dadosPCA,gs.best_estimator_)
    
print("\nNO PCA")
testMethod(dados,gs.best_estimator_)
```
